```python
# File: ReActAgent.py

class ReActAgent:
    def reason(self, query):
        print(f"Reasoning about: {query}")
        return "Let's search online for this information."

    def act(self):
        print("Using search tool...")
        return "The capital of France is Paris."

    def answer(self, query):
        reasoning = self.reason(query)
        print("Reasoning:", reasoning)
        result = self.act()
        return result

if __name__ == "__main__":
    agent = ReActAgent()
    print(agent.answer("What is the capital of France?"))
```

2. CodeAct Agent

```python
# File: CodeActAgent.py

class CodeActAgent:
    def generate_code(self, query):
        if "add" in query:
            return "result = 3 + 5"
        return "result = 0"

    def execute_code(self, code):
        local_vars = {}
        exec(code, {}, local_vars)
        return local_vars.get("result", None)

    def run(self, query):
        code = self.generate_code(query)
        print("Generated Code:", code)
        result = self.execute_code(code)
        return result

if __name__ == "__main__":
    agent = CodeActAgent()
    print("Code Result:", agent.run("add two numbers"))
```

ðŸ”§ 3. Modern Tool Use (MCP Pattern)

```python
# File: ModernToolAgent.py

class MCP:
    def search(self, query):
        return f"Kagi result for: {query}"

    def cloud_action(self):
        return "AWS S3 file uploaded."

class ModernToolAgent:
    def __init__(self):
        self.mcp = MCP()

    def handle_query(self, query):
        print("Using MCP to perform tasks...")
        search_result = self.mcp.search(query)
        cloud_result = self.mcp.cloud_action()
        return search_result, cloud_result

if __name__ == "__main__":
    agent = ModernToolAgent()
    search_out, cloud_out = agent.handle_query("agentic AI")
    print("Search Output:", search_out)
    print("Cloud Output:", cloud_out)
```

4. Self-Reflection Agent

```python
# File: SelfReflectAgent.py

class SelfReflectAgent:
    def generate(self, query):
        return "Pariss is the capital of France."

    def critique(self, text):
        if "Pariss" in text:
            return True, text.replace("Pariss", "Paris")
        return False, text

    def reflect(self, query):
        draft = self.generate(query)
        print("Initial Output:", draft)
        needs_fix, revised = self.critique(draft)
        if needs_fix:
            print("Critique Applied.")
        return revised

if __name__ == "__main__":
    agent = SelfReflectAgent()
    print("Final Output:", agent.reflect("What is the capital of France?"))
```

5. Multi-Agent Workflow

```python
# File: MultiAgentWorkflow.py

class SubAgent:
    def __init__(self, name):
        self.name = name

    def process(self, input_text):
        return f"{self.name} processed: {input_text}"

class AggregatorAgent:
    def __init__(self):
        self.agents = [SubAgent("Summarizer"), SubAgent("Translator")]

    def run(self, query):
        results = [agent.process(query) for agent in self.agents]
        return " | ".join(results)

if __name__ == "__main__":
    agent = AggregatorAgent()
    print("Aggregated Output:", agent.run("Translate and summarize this article"))
```

6. Agentic RAG (Retrieval-Augmented Generation)

```python
# File: AgenticRAG.py

class VectorDatabase:
    def __init__(self):
        self.data = {
            "RAG": "Retrieval-Augmented Generation uses vector search to add memory to LLMs."
        }

    def retrieve(self, query):
        return self.data.get(query, "No relevant context found.")

class RAGAgent:
    def __init__(self):
        self.vector_db = VectorDatabase()

    def answer(self, query):
        context = self.vector_db.retrieve(query)
        return f"Answer (using context): {context}"

if __name__ == "__main__":
    agent = RAGAgent()
    print(agent.answer("RAG"))
```
